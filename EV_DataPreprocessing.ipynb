# New Section

```python
# Tag Make and Model
Make = ['Chevrolet', 'Ford', 'Ford', 'GMC', 'Hyundai', 'Hyundai', 'Kia', 'Nissan', 'Subaru', 'Rivian','Tesla','Tesla', 'Volkswagen']
Model = ['Bolt', 'F150 Lightning', 'Mustang Mach-E', 'Hummer', 'Ioniq 5', 'Kona EV', 'EV6', 'Leaf', 'R1T', 'RAV4','Model 3', 'Model Y','ID']

def tag_model(brand):
  for tag in Model:
    if tag in brand:
      return tag
    else:
      pass

def tag_make(mak):
  for cm in Make:
    if cm in mak:
      return cm
    else:
      pass

mod = []
for i in range(len(all_SW_cmt['Purchased'])):
  brand = all_SW_cmt['Purchased'][i]
  mod.append(tag_model(brand))

all_SW_cmt.insert(0,'Model',mod)

car_make = []
for i in range(len(all_SW_cmt['Purchased'])):
  mak = all_SW_cmt['Purchased'][i]
  car_make.append(tag_make(mak))

all_SW_cmt.insert(0,'Make', car_make)
all_SW_cmt
```

# Reddit Clean

```python
# Remove url
url_cleared = []
for i in range(len(Red_concat)):
  result = re.sub(r'http\S+', '', Red_concat['Verbatim Text'][i])
  url_cleared.append(result)

Red_concat.insert(3, 'url_clean_text', url_cleared)
Red_concat = Red_concat.drop('Verbatim Text', axis=1)

Red_concat
```

```python
# Clear Emoji
emoji_cleared = []
for i in range(len(Red_concat)):
  text = clean(Red_concat['url_clean_text'][i], no_emoji=True)
  emoji_cleared.append(text)

Red_concat.insert(3,'emoji_clear_text', emoji_cleared)
Red_concat = Red_concat.drop('url_clean_text', axis=1)
Red_concat
```

```python
# Remove mentions and hashtag.
for text in Red_concat['emoji_clear_text']:
  text.replace('@', 'at')
  text = re.sub("#[A-Za-z0-9_]+","", text)
```

```python
Red_concat
```

```python
# capitalize the sentences (applying trasnformer does not need to be lowercased)
cap_text = []
for text in Red_concat['emoji_clear_text']:
  text = text.capitalize()
  cap_text.append(text)

Red_concat.insert(3,'Verbatim Text', cap_text)
Red_concat = Red_concat.drop('emoji_clear_text', axis=1)
Red_concat
```
